Defining function for maximum frequencies for each chunk.

# load_signal function

**The important default parameters are:**

• filename (compulsory parameter): This refers to the location of the .wav file.

• clip: This variable accepts a two-element list [*a*,*b*] and clips the input signal between the times a and b (in seconds) and returns the signal. This can be useful to remove the first *a* seconds or the last *b* seconds of your sound file. By default, no clipping is done.

• plot: This is a boolean variable that can either be True or False, based on whether one wants to plot the resulting signal or not. By default, the signal is not plotted.

**The function returns three values:**

• samplerate: This refers to the sample rate of the signal, defined as the number of data points of the input signal per second. Most recordings are taken at 44,100 Hz, which means that there are 44,100 samples in each second.

• times: An array of each individual time-point of the data.

• signal: An array containing the value of the amplitude of the signal at each time-step in times.

# power_spectrum function

**The important default parameters are:**

• chunk (compulsory parameter): An array of amplitude values, usually a specific chunk that was returned by the chunk_signal function.

• samplerate (compulsory parameter)

• lowx and highx: The range of frequencies within which to search for a maximum.

**The function returns:**

• maxfreq: The frequency in a particular range that contributes the greatest to the signal in a given range of frequencies.

```python
def maxim(file, cs):
  global sr, dominantfrequencies, averagetime, maxfreq, minfreq
  sr, times, sig = dop.load_signal(file, clip = [0,35])
  averagetime, chunks = dop.chunk_signal(times, sig, size = cs)

  dominantfrequencies = np.zeros_like(averagetime)
  for i in range(0,len(averagetime)): 
    dominantfrequencies[i] = dop.power_spectrum(chunks[i], sr, lowx = 2000, highx = 4100) # Finding dominant frequency for each chunk
  plt.plot(averagetime, dominantfrequencies)
  plt.xlim(1,15)
  maxfreq = max(dominantfrequencies)
  minfreq = min(dominantfrequencies)
  plt.axhline(maxfreq, color = 'green')
  plt.axhline(minfreq, color = 'green')
  plt.show()
```

Defining function that returns the speed of sound based on equation (2)

```python
def speed(v, maxfreq, minfreq):
  v_0 = 4000
  C_s = (2*v*v_0)/(maxfreq-minfreq)
  return C_s
velocity_list = [] # Storing all the velocities
soundspeed = [] # Storing all the values of speed of sound
```

Plotting dominant frequencies against average time for chunks for each reading. The speed of sound is also calculated for each reading.

```python
print('Average time vs dominant frequencies for the first reading')
maxim("/content/drive/MyDrive/Doppler/Copy of 1_First_reading_slowest.wav", 32050)
v = (2*np.pi*0.45)/4 # Eyeballing the value of time period
velocity_list.append(v)
C_s = speed(v, maxfreq, minfreq)
soundspeed.append(C_s)
print('Speed of sound:', C_s)
```
